```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import optimize
temp_max=np.array([39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min=np.array([21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
months=np.arange(12)
plt.plot(months,temp_max,'ro')
plt.plot(months,temp_min,'bo')
plt.xlabel('Month')
plt.ylabel('Min and Max Temperature')

def yearly_Temperature(times,avg,ampl,time_offset):
    return (avg+ampl*np.cos((times+time_offset)*2*np.pi/times.max()))
res_max,cov_max=optimize.curve_fit(yearly_Temperature,months,temp_max)
res_min,cov_min=optimize.curve_fit(yearly_Temperature,months,temp_min)
days=np.linspace(0,12,365)
plt.figure()
plt.plot(months,temp_max,'ro')
plt.plot(days,yearly_Temperature(days,*res_max),'r-')
plt.plot(months,temp_min,'bo')
plt.plot(days,yearly_Temperature(days,*res_min),'b-')
plt.xlabel('Month')
plt.ylabel('Temperature($^\circ$C)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

url='https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv'
df = pd.read_csv(url)
df['sex']=df['sex'].fillna(value='male')
fig1,ax1 = plt.subplots()
labels='Male','Female'
dfs=list(df['sex'].value_counts())
ax1.pie(dfs,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```
